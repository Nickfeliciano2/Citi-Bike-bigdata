# Glob merge

```python
#glob csv files

path = 'data'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    raw_df = pd.read_csv(filename, index_col=None, header=0)
    li.append(raw_df)

frame = pd.concat(li, axis=0, ignore_index=True)
```
